```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = np.loadtxt("cg_results.txt", skiprows=1)
Ns, iterations, times = data[:, 0], data[:, 1], data[:, 2]
```

```python
# Plot CG Iterations vs N
plt.figure()
plt.plot(Ns, iterations, marker='o')
plt.title("CG Iterations vs N")
plt.xlabel("N")
plt.ylabel("Iterations")
plt.grid(True)
plt.show()
plt.savefig("cg_iterations.png")
```

```python
# Plot time vs N
plt.figure()
plt.plot(Ns, times, marker='o')
plt.title("CG Time vs N")
plt.xlabel("N")
plt.ylabel("Time (s)")
plt.grid(True)
plt.show()
plt.savefig("cg_time.png")
```

```python
N = 256  
solution = np.loadtxt(f"solution_{N}.txt")

x = np.linspace(0, 1, N)
y = np.linspace(0, 1, N)
X, Y = np.meshgrid(x, y)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, solution, cmap='viridis')

ax.set_title(f"3D Solution Surface for N = {N}")
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("u(x1, x2)")

plt.tight_layout()
plt.show()
plt.savefig(f"solution_surface_N{N}.png")
```
